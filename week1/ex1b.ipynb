## Ex1 Loading data

```python
import numpy as np
iris_data = np.loadtxt("iris_data/irisdata.txt", comments="%")
iris_data
```

```python
x = iris_data[:50, :4]
n_feat = x.shape[1]
n_obs = x.shape[0]
x.shape, n_feat, n_obs
```

## Ex2 EDA

```python
# Look at indiv features
sep_l = x[:, 0]
sep_w = x[:, 1]
pet_l = x[:, 2]
pet_w = x[:, 3]
sep_l.shape, sep_w.shape, pet_l.shape, pet_w.shape
```

### Calculate variances in each feature

```python
# Use ddof = 1 to make an unbiased estimate
var_sep_l = sep_l.var(ddof=1)
var_sep_w = sep_w.var(ddof=1)
var_pet_l = pet_l.var(ddof=1)
var_pet_w = pet_w.var(ddof=1)
var_sep_l, var_sep_w, var_pet_l, var_pet_w
```

## Ex3 Calculate covariance

```python
# Obtain covariance between sepal length and petal length
N = sep_l.size
cov_length = (1.0 / (N - 1)) * (sep_l * pet_l).sum()
cov_width = (1.0 / (N - 1)) * (sep_w * pet_w).sum()
cov_length, cov_width
```

```python
x = np.stack((sep_l, pet_l), axis=0)
x, np.cov(x), np.cov(sep_l, pet_l), np.cov(sep_w, pet_w)
```

## Ex 5 PCA

```python
mn = np.mean(x, axis=0)
data = x - mn
data
```

```python
# Need to transpose data as the function treats rows as the features
# marray_like
# A 1-D or 2-D array containing multiple variables and observations. 
# Each row of m represents a variable, and each column a single observation of all those variables. Also see rowvar below.
cov = np.cov(data.T)
cov
```

```python
size = data.shape[0]
covariance = (1.0 / (size - 1)) * data.T @ data
covariance
```

## Ex6 Principal components 

```python
values, vectors = np.linalg.eig(covariance)
values, vectors
```

## Ex7 Plot

```python
v_norm = values / values.sum() * 100
plt.plot(v_norm)
plt.xlabel("Principal component")
plt.ylabel("Percent explained variance")
plt.ylim([0, 100])

plt.show()
```

```python
pc_proj = vectors.T.dot(data.T)
pc_proj.shape
d = pd.DataFrame(pc_proj.T, columns=['Sepal length', 'Sepal width', 'Petal length', 'Petal width'])
plt.figure()  # Added this to make sure that the figure appear
sns.pairplot(d)
plt.show()
```

```python
from sklearn import decomposition
pca = decomposition.PCA()
pca.fit(x)
values_pca = pca.explained_variance_
exp_var_ratio = pca.explained_variance_ratio_
vectors_pca = pca.components_

data_transform = pca.transform(x)
data_transform.shape
```

```python
d = pd.DataFrame(
    data_transform, columns=["Sepal length", "Sepal width", "Petal length", "Petal width"]
)
plt.figure()  # Added this to make sure that the figure appear
sns.pairplot(d)
plt.show()
```
